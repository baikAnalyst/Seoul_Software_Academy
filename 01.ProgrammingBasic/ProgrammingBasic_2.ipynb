## break, continue
for, while에서 제어흐름을 벗어나기 위해 사용
- break: 조건에 만족하면 반복을 완전중단
- continue: 조건에 만족하면 이번 반복만 중단하고(출력하지 않고) 처음으로 돌아가 다음 반복 수행

```python
# break
for i in range(5):
  if(i == 3):
    break
  print(i, end='')
```

```python
# continue
for i in range(5):
  if (i == 3):
    continue
  print(i, end='')
```

## 실습 02

가변인수와 고정인수를 사용해 모든값을 더하거나 곱하는 함수 작성


```python
def calc(operator, * numbers):
  if operator == '+':
      result = 0
      for num in numbers:
          result += num
      return result

  elif operator == '*':
      result = 1
      for num in numbers:
          result *= num
      return result
```

```python
calc("+", 1,2,3,4,5)
```

```python
calc('*', 1,2,3,4,5)
```

### factorial by recursive function

```python
def factorial(n):
  if n == 1:
    return 1
  return n * factorial(n-1)

factorial(5)
```

```python
factorial(0)
```

```python
def factorial(n):
  if n == 1:                      # n 이 1이면
    return 1                       # 1을 반환하고 호출을 끝냄
  return n * factorial(n-1)  # n이 1이 될 때까지 함수 내부에서 자기 자신을 또 다시 호출하기를 반복

factorial(5)

# n * factorial(n-1)
# n * (n-1) * factorial (n-2)
# n * (n-1) * (n-2) * factorial (n-3)
# n * (n-1) * (n-2) * (n-3) * factorial(n-4)
# n * (n-1) * (n-2) * (n-3) * (n-4) * factorial(n-5)
# ...
# n * (n-1) * (n-2) * (n-3) * ... * 1
```

### map - process iterable data type values using function


#### map function
- map(function, variable(which is iterable such as sequential data type))

```python
# general function
a =list(range(1, 11))

def f(x):
  if x % 2 == 0:
    return 0
  else:
    return x

list(map(f, a))
```

```python
# if conditional in one-line - True
'if True 인 경우의 값' if True else "if false인 경우의 값"
```

```python
# if conditional in one-line - False
'if True인 경우의 값' if False else "if False인 경우의 값"
```

```python
# lambda expression
a = list(range(1, 11))

list(map(lambda x: 0 if x % 2==0 else x, a))
```

#### Not able to use elif in lambda expression, use if and else instead to make the same result

```python
a = list(range(1, 11))

# (if) 2의 배수인 경우 0, (elif) 2의 배수가 아니면서 3의 배수인 경우 1, (else) 그 외인 경우 2
list(map(lambda x : 0 if x%2 == 0 else 1 if x % 3 == 0 else 2, a))
```

#### filter  - filter only (boolean) True values by function from iterable data type
- filter(f, a)

```python
a = [2, 6, 4, 3, 6, 8, 3, 9, 6]

list(filter(lambda x : x>2 and x < 8, a)) # list a에서 2보다 크고 8보다 작은 값이라서 True인 것만 걸러냄
```
