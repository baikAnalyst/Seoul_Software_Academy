```python
matrix_1 = [[1, 2],[3, 4]]
matrix_2 = [[5, 6],[7, 8]]
```

```python
# 행렬합
matrix_result = []
for i in range(len(matrix_1)):
  tmp = []
  for j in range(len(matrix_2)):
    tmp.append(matrix_1[i][j] + matrix_2[i][j])
  matrix_result.append(tmp)

print(matrix_result)
```

```python
import numpy as np

matrix_result = np.array(matrix_1) + np.array(matrix_2)
print(matrix_result)
```

```python
# 행렬에 숫자 1씩 더하기

for i in range(len(matrix_1)):
    for j in range(len(matrix_2)):
        matrix_1[i][j] = matrix_1[i][j] + 1
print(matrix_1)
```

```python
# 브로드캐스트 - 형태가 다른 형렬끼리 계산할 수 있게

matrix_1 = np.array(matrix_1) + 1
matrix_1
```

- random()   -> random.random() : 0이상 1미만의 실수 난수(랜덤한 수) 생성
- randint()  -> random.randint(start, stop): start이상 stop이하 범위의 정수 난수 생성
- randrange()-> random.randrange(low, high, step): low이상 high이하 범위에서 low에 step을 더해 간 범위 내에서 정수 난수 생성(*일련의 규칙성을 갖는 수열 중, 무작위추출을 한다고 볼 수도 있음)<br>
 https://m.blog.naver.com/regenesis90/222343835845

- np.random.seed seed를 통한 난수 생성

- np.random.randint 균일 분포의 정수 난수 1개 생성
- np.random.rand 0부터 1사이의 균일 분포에서 난수 matrix array생성
- np.random.randn 가우시안 표준 정규 분포에서 난수 matrix array생성

- np.random.shuffle 기존의 데이터의 순서 바꾸기
- np.random.choice 기존의 데이터에서 sampling (주어진 리스트 안에 있는 숫자 뽑기)
- np.unique 데이터에서 중복된 값을 제거하고 중복되지 않는 값의 리스트를 출력
- np.bincount 발생하지 않은 사건에 대해서도 카운트를 해준다

```python
def Age_Range(age):
  if age // 10 == 0:
    return '유아'
  elif age // 10 == 1:
    return '10대'
  elif age // 10 == 2:
    return '20대'
  elif age // 10 == 3:
    return '30대'
  elif age // 10 == 4:
    return '40대'
  elif age // 10 == 5:
    return '50대'
  elif age // 10 == 6:
    return '60대'
  else:
    return '노인'
```

```python
titanic['연령대'] = titanic['Age'].apply(Age_Range)
titanic.head()
```
