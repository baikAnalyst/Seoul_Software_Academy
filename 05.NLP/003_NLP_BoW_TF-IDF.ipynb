# TF-IDF
: Term Frequency - Inverse Document Frequency



```python
docs = ['오늘 동물원에서 원숭이와 코끼리를 봤어',
            '동물원에서 원숭이에게 바나나를 줬어 바나나를']
```

```python
# 1. tokenize by space

doc_ls = [doc.split(' ') for doc in docs]
doc_ls
```

```python
# 2. unique tokens dictionary
#   2-1. generate empy dictionary

from collections import defaultdict

word2id = defaultdict(lambda : len(word2id))
word2id
```

```python
#   2-2. imputation: unique tokens into empty dictionary which generated above

for doc in doc_ls:
    for token in doc:
        word2id[token]
        print(token)
        print('\t', word2id)
word2id
```

```python
# 3. TF computation
#     TF = frequency of specific words' appearance / frequency of total words' appearance in documents

import numpy as np
                          # row           # col  ->  8x2 matrix
TDM = np.zeros((len(word2id), len(doc_ls)), dtype=int)

for i, doc in enumerate(doc_ls):
    print(doc)
    for token in doc:
        TDM[word2id[token], i] += 1     # token index in column
        print(token)
        print(TDM)
        print('\t')
```

```python
# TF computation

def computeTF(TDM):
    doc_len = len(docs)        # num of docs:2
    word_len = len(word2id) # num of unique tokens:8
    print('문서수 : ', doc_len)
    print('전체 단어수: ', word_len)

    tf = np.zeros((word_len, doc_len))
    print(tf)

    for doc_i in range(len(doc_ls)) :
        print(docs[doc_i])
        for word_i in range(len(word2id)):
            tf[word_i, doc_i] = TDM[word_i, doc_i]/TDM[:, doc_i].sum()  # TF computation
        print(tf)
    return tf
```

```python
tf = computeTF(TDM)
```

```python
tf
```

```python
# 4. IDF computation
#     IDF = -log10(number of documents that words have appeared / total documents)

import math

def computeIDF(TDM) :
    doc_len = len(docs)  # num of docs:2
    word_len = len(word2id)
    print('문서수 : ', doc_len)
    print('전체 단어수: ', word_len)

    idf = np.zeros(word_len)
    print(idf)

    for doc_i in range(len(doc_ls)) :
        for word_i in range(len(word2id)) :  #np.count_nonzero: to count words appeared
            idf[word_i] = -math.log10(np.count_nonzero(TDM[word_i, :]) / doc_len)
    print(idf)
    return idf

# https://mathbang.net/596#gsc.tab=0
```

```python
idf = computeIDF(TDM)
idf
```

```python
# 5. TF-IDF computation: TF x IDF

def computeTFIDF(TDM):
    tf = computeTF(TDM)
    idf = computeIDF(TDM).T   # inverse df -> Transpose
    tfidf = np.zeros(tf.shape)
    print( )

    for doc_i in range(len(doc_ls)) :
        print(docs[doc_i])
        for word_i in range(len(word2id)) :
            tfidf[word_i, doc_i] = tf[word_i, doc_i] * idf[word_i]

    print('TF-IDF 최종')
    print(tfidf)
    return tfidf
```

```python
tfidf = computeTFIDF(TDM)
```

```python
tfidf
```

```python
# 6. visualization

import pandas as pd

sorted_vocab = sorted((value, key) for key, value in word2id.items())
vocab = [v[1] for v in sorted_vocab]
print('vocab : ', vocab)
tfidf = computeTFIDF(TDM)
pd.DataFrame(tfidf.T, columns=vocab)
```

```python
# sklearn

docs = ['오늘 동물원에서 원숭이와 코끼리를 봤어',
            '동물원에서 원숭이에게 바나나를 줬어 바나나를']

from sklearn.feature_extraction.text import TfidfVectorizer
tfidfV= TfidfVectorizer()
tfidfV = tfidfV.fit(docs)                              # .fit(): docs -> BoW -> tf-idf compuation for each terms -> vectorize terms in each term's index
tfidfV.transform(docs).toarray()                 # .transform(): corpus to list to iterate / .toarray(): list to numpy array
vocab = tfidfV.get_feature_names_out()
```

```python
import pandas as pd

df = pd.DataFrame(tfidfV.transform(docs).toarray(), columns = vocab)
df
```
